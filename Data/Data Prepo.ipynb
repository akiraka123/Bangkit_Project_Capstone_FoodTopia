```python
data = data.drop(columns=['AuthorId', 'AuthorName', 'CookTime', 'PrepTime', 'TotalTime', 'DatePublished', 'Description', 'AggregatedRating', 'ReviewCount', 'RecipeYield'])
```

```python
data.dropna(how = 'any', inplace = True)
data.isnull().sum()
```

```python
data.describe()
```

```python
# Get indices of rows where 'column_name' equals 'value_to_drop'
indices_to_drop = data[data['Images'] == 'character(0)'].index

# Drop the rows by index
data = data.drop(indices_to_drop)
# Display unique values in a specific column
unique_values = data['Images'].unique()
unique_values

```

```python
data_clean = data.to_excel('/recipe_clean.xlsx')
```

```python
data.to_csv('foodcom-recipes-and-reviews/cleaned_data.csv', index=False)
```
